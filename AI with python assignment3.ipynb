1- Write a `for` loop which repeats the steps 1)-3) below for values of `n` ranging as

500, 1000, 2000, 5000, 10000, 15000, 20000, 50000, 100000

2- Use numpy to simulate throwing of two dice n` times. Compute the sum of the dice.

3- Use numpys `histogram()` function to compute the frequencies as

`h,h2 = np.histogram(s,range(2,14))`
where s contains the sum.

4- Use matplotlib's bar function to plot the histogram as

`plt.bar(h2[:-1],h/n)`

and show the value of `n` in the title.

5- What do you observe? You may need to run the loop a few times to see it.

6- How is this related to "regression to the mean"?

```python
import numpy as np
import matplotlib.pyplot as plt

# Values of n
n_values = [500, 1000, 2000, 5000, 10000, 15000, 20000, 50000, 100000]

# Theoretical probabilities for sums 2 to 12
theoretical_probabilities = {
    2: 1/36, 3: 2/36, 4: 3/36, 5: 4/36, 6: 5/36,
    7: 6/36, 8: 5/36, 9: 4/36, 10: 3/36, 11: 2/36, 12: 1/36
}

# Loop through each value of n
for n in n_values:
    # Step 1: Simulate n rolls of two dice
    die1 = np.random.randint(1, 7, n)  # Roll for die 1
    die2 = np.random.randint(1, 7, n)  # Roll for die 2
    sums = die1 + die2                # Sum of both dice
    
    # Step 2: Calculate frequencies of each sum
    counts = {i: np.sum(sums == i) for i in range(2, 13)}
    
    # Step 3: Calculate relative frequencies
    probabilities = {k: v / n for k, v in counts.items()}
    
    # Step 4: Plotting the histogram of relative frequencies
    plt.figure(figsize=(10, 6))
    plt.bar(probabilities.keys(), probabilities.values(), color='blue', alpha=0.6, label='Simulated')
    plt.plot(theoretical_probabilities.keys(), theoretical_probabilities.values(), 'ro-', label='Theoretical')
    plt.xlabel('Sum of Two Dice')
    plt.ylabel('Probability')
    plt.title(f'Probability Distribution of Sum of Two Dice (n = {n})')
    plt.legend()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the number of times to throw the dice
n = 10000  # You can change this as needed

# Step 1: Simulate n rolls of two dice and compute the sum of each roll
die1 = np.random.randint(1, 7, n)  # Roll for die 1
die2 = np.random.randint(1, 7, n)  # Roll for die 2
s = die1 + die2                    # Sum of both dice

# Step 2: Compute the frequencies of each sum
h, h2 = np.histogram(s, bins=range(2, 14))

# Step 3: Plot the histogram using plt.bar
plt.figure(figsize=(10, 6))
plt.bar(h2[:-1], h / n, width=0.8, color='blue', alpha=0.7)  # Plot relative frequencies
plt.xlabel('Sum of Two Dice')
plt.ylabel('Probability')
plt.title('Probability Distribution of the Sum of Two Dice')
plt.xticks(range(2, 13))  # Set x-axis ticks to match the sums 2 to 12
plt.show()
```

Consider the data from the file weight-height.csv.

1- Inspect the dependence between height and weight using a scatter plot. You may use either of the variables as independent variable.

2- Choose appropriate model for the dependence

3- Perform regression on the data using your model of choice

4- Plot the results

5- Compute RMSE and R2 value

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Load the data
data = pd.read_csv('weight-height.csv')

# Step 1: Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data['Height'], data['Weight'], alpha=0.5, color='blue')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Scatter Plot of Height vs. Weight')
plt.show()

# Step 2-3: Fit linear regression
X = data[['Height']].values
y = data['Weight'].values
model = LinearRegression()
model.fit(X, y)

# Step 4: Plot regression line
y_pred = model.predict(X)
plt.figure(figsize=(10, 6))
plt.scatter(data['Height'], data['Weight'], alpha=0.5, color='blue', label='Data points')
plt.plot(data['Height'], y_pred, color='red', label='Regression Line')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs. Weight with Regression Line')
plt.legend()
plt.show()

# Step 5: Calculate RMSE and R
rmse = np.sqrt(mean_squared_error(y, y_pred))
r2 = r2_score(y, y_pred)
print("Root Mean Squared Error (RMSE):", rmse)
print("R Score:", r2)
```
